# celltrion beauty mall
- best 상품 데이터 40개 수집
- https://www.celltrionbeauty.com/main/best-items
- 상품 이미지 40개 수집

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

## 1. URL

```python
# find url
url ='https://www.celltrionbeauty.com/main/best-items'
```

## 2. Request > Response : HTML(str)

```python
# request > response
response = requests.get(url)
response
```

## 3. HTML(str) > BeautifulSoup Object : CSS Selector > DataFrame

```python
# make BeautifulSoup Object
dom = BeautifulSoup(response.text, 'html.parser')
type(dom)
```

```python
# find item list css selector
selector = '#productListBody > div' # 목록 데이터들 (40개) 영역을 CSS Selector로 가져옴 
elements = dom.select(selector)
len(elements)
```

```python
# find item css selector
element = elements[0]
data = {
    'title' : element.select_one('.tit').text, # 제목 선택
    'price' : element.select_one('.pri').text, # 가격 선택
    'img'   : element.select_one('img').get('data-src'), #이미지 가져옴 , 이미지는 src= 속성 값을 가져와야함 : get('data-src')
    
    
    
}
#element 확인  
data
```

```python
# parsing all elements
items = []
for element in elements:
    data = {
    'title' : element.select_one('.tit').text, # 제목 선택
    'price' : element.select_one('.pri').text, # 가격 선택
    'img'   : element.select_one('img').get('data-src'), #이미지 가져옴 , 이미지는 src= 속성 값을 가져와야함 : get('data-src')
    }
    items.append(data)
items[:3]

```

```python
df = pd.DataFrame(items)
df.tail(2)
```

## 4. Download Images

```python
# 파일 시스템다룸 : 디렉 생성 
import os
# 이미지 전처리 
from PIL import Image as pil
```

```python
# make directory
path = 'data'
os.makedirs(path, exist_ok = True)
```

```python
os.listdir()
```

```python
# download image
link = df.loc[0, 'img']
link
```

```python
filename = 'test'
filelink = f'{path}/{filename}.jpg'

response = requests.get(link)

with open(filelink, 'wb') as file: # 파일을 바이너리 모드로 열어서('wb') HTTP 응답의 내용을 파일로 저장.
    file.write(response.content)
```
