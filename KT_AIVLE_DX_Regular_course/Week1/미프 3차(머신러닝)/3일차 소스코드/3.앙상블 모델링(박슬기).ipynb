### (1) 라이브러리 로딩

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.auto import tqdm
import warnings
warnings.filterwarnings('ignore')

import joblib
from sklearn.ensemble import StackingClassifier, VotingClassifier

# 필요한 라이브러리, 함수 로딩 ------------------
from sklearn.model_selection import train_test_split
from sklearn.metrics import *
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from lightgbm import LGBMClassifier
```

* 제공 함수 생성
    * 변수 중요도를 시각화할 수 있는 함수를 제공합니다.
    * 입력 :
        * importance : 트리모델의 변수 중요도(예: model.feature_importances_)
        * names : 변수 이름 목록(예 : x_train.columns
        * result_only  : 변수 중요도 순으로 데이터프레임만 return할지, 그래프도 포함할지 결정. False이면 결과 데이터프레임 + 그래프
        * topn : 중요도 상위 n개만 표시. all 이면 전체.
    * 출력 :
        * 중요도 그래프 : 중요도 내림차순으로 정렬
        * 중요도 데이터프레임 : 중요도 내림차순으로 정렬

```python
# 변수의 특성 중요도 계산하기
def plot_feature_importance(importance, names, result_only = False, topn = 'all'):
    feature_importance = np.array(importance)
    feature_name = np.array(names)

    data={'feature_name':feature_name,'feature_importance':feature_importance}
    fi_temp = pd.DataFrame(data)

    #변수의 특성 중요도 순으로 정렬하기
    fi_temp.sort_values(by=['feature_importance'], ascending=False,inplace=True)
    fi_temp.reset_index(drop=True, inplace = True)

    if topn == 'all' :
        fi_df = fi_temp.copy()
    else :
        fi_df = fi_temp.iloc[:topn]

    #변수의 특성 중요도 그래프로 그리기
    if result_only == False :
        plt.figure(figsize=(10,20))
        sns.barplot(x='feature_importance', y='feature_name', data = fi_df)

        plt.xlabel('importance')
        plt.ylabel('feature name')
        plt.grid()

    return fi_df
```

## **2. 데이터 전처리**

* 가변수화, 데이터 분할, NaN 확인 및 조치, 스케일링 등 필요한 전처리를 수행한다.


### (1) 데이터 분할1 : x, y

* 세부 요구사항
    - x, y로 분할합니다.

```python
drop_cols ='Activity'
x = data.drop(columns = drop_cols)
y = data.loc[:,drop_cols]

x_test = test.drop(columns = drop_cols)
y_test= test.loc[:,drop_cols]
```

### (2) 데이터분할2 : train, validation

* 세부 요구사항
    - train : val = 8 : 2 혹은 7 : 3
    - random_state 옵션을 사용하여 다른 모델과 비교를 위해 성능이 재현되도록 합니다.

```python
x_train , x_val , y_train , y_val = train_test_split(x,y,test_size = 0.3)
```

### (3) 스케일링


* 세부 요구사항
    - 스케일링을 필요로 하는 알고리즘 사용을 위해서 코드 수행
    - min-max 방식 혹은 standard 방식 중 한가지 사용.

```python
scaler = MinMaxScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_val = scaler.transform(x_val)
x_test = scaler.transform(x_test)
```

### (4) (옵션)KNN을 위한 전처리
* KNN은 대량의 데이터에 대한 거리 계산을 할때, 데이터를 연속적인 메모리 블록에 저장하는 것이 중요합니다.
* c_contiguous 속성을 요구

```python
x_train = np.ascontiguousarray(x_train)
x_val = np.ascontiguousarray(x_val)
```

### (5) (옵션) Activity 범주 0~5 변경

```python
activity_mapping = {
    'LAYING': 0,
    'STANDING': 1,
    'SITTING': 2,
    'WALKING': 3,
    'WALKING_UPSTAIRS': 4,
    'WALKING_DOWNSTAIRS': 5
}
y_train = y_train.map(activity_mapping)
y_val = y_val.map(activity_mapping)
y_test = y_test.map(activity_mapping)
```

## **3.앙상블 모델링**
* Voting 혹은 Stacking 방식의 앙상블 모델을 구성하고 성능을 평가해 봅시다.


### (1) 모델 로딩하기

* 세부 요구사항
    * 2.기본 모델링 파일에서 저장한 모델을 로딩합니다.(joblib.load)


```python
import joblib

model_knn = joblib.load('knn.pkl')
model_lgbm = joblib.load('LGBM.pkl')
model_lr = joblib.load('lr.pkl')
model_svm = joblib.load('svm.pkl')
```

### (2) 앙상블 모델 생성


#### Voting

```python
voting_clf = VotingClassifier(estimators=[
    ('knn', model_knn), 
    ('svm', model_svm), 
    ('lr', model_lr), 
    ('lgbm', model_lgbm)], voting='hard', weights=[4, 1, 2, 3])  # 가중치(전 파일에서 정확도가 더 잘 나온 순으로 지정했습니다!)
```

```python
# 모델 학습
voting_clf.fit(x_train, y_train)
```

```python
# 평가
y_pred = voting_clf.predict(x_test)

# 결과
print(confusion_matrix(y_test, y_pred))
print()
print('Classification Report  \n' , classification_report(y_test, y_pred))

report = classification_report(y_test, y_pred, output_dict=True)
accuracy = report['accuracy']
print(f'Accuracy: {accuracy:.4f}')
```

#### Stacking

##### Logistic Regression: 간단하고 빠름

```python
# 최종 모델 Logistic Regression
stacking_clf_lr = StackingClassifier(estimators=[
    ('knn', model_knn), 
    ('svm', model_svm), 
    ('lr', model_lr), 
    ('lgbm', model_lgbm)],
    final_estimator=LogisticRegression(), cv=5
)
```

```python
# 모델 학습
stacking_clf_lr.fit(x_train, y_train)
```

```python
# 평가
y_pred_stacking_lr = stacking_clf_lr.predict(x_test)

# 결과
print(confusion_matrix(y_test, y_pred_stacking_lr))
print()
print('Classification Report  \n' , classification_report(y_test, y_pred_stacking_lr))

report_lr = classification_report(y_test, y_pred_stacking_lr, output_dict=True)
accuracy_lr = report_lr['accuracy']
print(f'Accuracy: {accuracy_lr:.4f}')
```

##### Random Forest: 복잡한 의사결정을 다루고, 다양한 특성 처리에 용이

```python
# 최종 모델 Random Forest 사용
stacking_clf_rf = StackingClassifier(estimators=[
        ('knn', model_knn), 
        ('svm', model_svm), 
        ('lr', model_lr), 
        ('lgbm', model_lgbm)],
    final_estimator=RandomForestClassifier(), cv=5
)
```

```python
# 모델 학습
stacking_clf_rf.fit(x_train, y_train)
```

```python
# 평가
y_pred_stacking_rf = stacking_clf_rf.predict(x_test)

# 결과
print(confusion_matrix(y_test, y_pred_stacking_rf))
print()
print('Classification Report  \n' , classification_report(y_test, y_pred_stacking_rf))

report_rf = classification_report(y_test, y_pred_stacking_rf, output_dict=True)
accuracy_rf = report_rf['accuracy']
print(f'Accuracy: {accuracy_rf:.4f}')
```

##### SVM:고차원 데이터에 적합

```python
# 최종 모델 SVM 사용
stacking_clf_svm = StackingClassifier(estimators=[
        ('knn', model_knn), 
        ('svm', model_svm), 
        ('lr', model_lr), 
        ('lgbm', model_lgbm)],
    final_estimator=SVC(), cv=5
)
```

```python
# 모델 학습
stacking_clf_svm.fit(x_train, y_train)
```

```python
# 평가
y_pred_stacking_svm = stacking_clf_svm.predict(x_test)

# 결과
print(confusion_matrix(y_test, y_pred_stacking_svm))
print()
print('Classification Report  \n' , classification_report(y_test, y_pred_stacking_svm))

report_svm = classification_report(y_test, y_pred_stacking_svm, output_dict=True)
accuracy_svm = report_svm['accuracy']
print(f'Accuracy: {accuracy_svm:.4f}')
```

##### LGBM: 복잡한 데이터에 적합

```python
# 최종 모델 LightGBM사용
stacking_clf_lgbm = StackingClassifier(estimators=[
        ('knn', model_knn), 
        ('svm', model_svm), 
        ('lr', model_lr), 
        ('lgbm', model_lgbm)],
    final_estimator=LGBMClassifier(), cv=5
)
```

```python
# 모델 학습
stacking_clf_lgbm.fit(x_train, y_train)
```

```python
# 평가
y_pred_stacking_lgbm = stacking_clf_lgbm.predict(x_test)

# 결과
print(confusion_matrix(y_test, y_pred_stacking_lgbm))
print()
print('Classification Report  \n' , classification_report(y_test, y_pred_stacking_lgbm))

report_lgbm = classification_report(y_test, y_pred_stacking_lgbm, output_dict=True)
accuracy_lgbm = report_lgbm['accuracy']
print(f'Accuracy: {accuracy_lgbm:.4f}')
```
