# merge 훈련

```python
# 라이브러리 불러오기 
import pandas as pd
import numpy as np
```

```python
# 데이터 읽어오기  
# Employee 
employee = pd.read_csv('https://bit.ly/HRDBEmployee')
# Department 
department = pd.read_csv('https://bit.ly/HRDBDepartment')
# Vacation 
vacation = pd.read_csv('https://bit.ly/HRDBVacation')
```

```python
display(employee.head())
display(department.head())
display(vacation.head())
```

## 분석
- employee 와 department 는 공통된 DeptID를 가지고 있음 , 이를 활용해 Join
- employee와 vacation은 공통된 EmpID 를 가지고 있음 이를 활용해 Join 

```python
# emp 데이터프레임 생성
my_cols = ['EmpID','EmpName','DeptID']

emp = employee.loc[:,my_cols].copy()

# 확인 
emp
```

```python
# dept 데이터프레임 
dept_cols = ['DeptID','DeptName']
dept = department.loc[:,dept_cols].copy()

#확인
dept
```

```python
# Join 병합
emp01 = pd.merge(emp, dept , how= 'inner' ,on ='DeptID') # merge는 2개 요소밖에 결합 못함 => 리스트 x 

# 컬럼 명이 일치하진 않지만 의미가 같은 경우   // 추천하진 않음 , 컬럼명을 일치하게 두는게 더 좋음 
#  emp01 = pd.merge(emp, dept , how= 'inner' ,left_on ='DeptID',right_on='DeptID')
emp01
```

```python
# Join 병합 right
emp01 = pd.merge(emp, dept , how= 'right' ,on ='DeptID') # merge는 2개 요소밖에 결합 못함 => 리스트 x 

emp01 # 홍보팀은 아직 직원채용이 되지 않아서 Employee 데이터프레임엔 없으나 Department 데이터프레임에는 존재함 
```
