문제설명
화가인 지연이는 그림판에서 N x M 크기의 격자판을 칠하여 그림을 그렸습니다. 각 칸은 검정색, 빨간색, 파란색 3개의 색 중 하나로 칠해져 있거나 빈 칸입니다. 그림은 얼핏 보면 검정색으로 구분된 여러 개의 영역으로 나뉜 것 같아 보입니다. 이때 검정색이 아닌 두 칸이 상하좌우로 인접하다면 하나의 영역에 속하는 것으로 간주합니다.

그림을 유심히 보던 지연이는 같은 영역 안에 있는 색들을 하나의 색으로 통일하고 싶었습니다. 그래서 다음과 같이 각각의 영역을 수정하기로 했습니다.
하나의 영역에 대해,
i) 빨간색과 파란색으로 칠해져 있는 칸의 수가 동일하거나, 파란색으로 칠해져 있는 칸이 더 많다면 빨간색으로 칠해져 있는 칸의 색깔을 지우고 빈 칸으로 만듭니다.
ii) 빨간색으로 칠해져 있는 칸이 더 많다면 이번에도 역시 파란색으로 칠해져 있는 칸의 색깔을 지우고 빈 칸으로 만듭니다.

그림의 정보가 주어질 때, 지연이가 수정 작업을 끝내고 그림에 남은 빨간색 칸, 파란색 칸의 개수를 각각 출력하세요.

```python
from collections import deque

def BFS(x,y, mmap,visited,N,M):
    
    dx = (0,1,0,-1)
    dy = (1,0,-1,0)
    queue = deque()
    queue.append((x,y))
    visited[x][y]= True
    
    # 빨간색, 파란색 어떤게 더 많을까? 
    reds = 0
    blues = 0
    if mmap[x][y] == 'A':
        reds +=1
    elif mmap[x][y] == 'B': # 변경
        blues+=1
    
    # 방문한 좌표를 따로 저장할 리스트
    visited_cells = []
    visited_cells.append((x,y))
    while queue:
        nx , ny = queue.popleft()
        
        for i in range(4):
            mov_x = nx +dx[i]
            mov_y = ny +dy[i]
            if 0<= mov_x <N and 0<=mov_y <M and visited[mov_x][mov_y]==False and mmap[mov_x][mov_y] in ('A','B'):
                queue.append((mov_x,mov_y))
                visited[mov_x][mov_y] =True
                if mmap[mov_x][mov_y] == 'A':
                    reds +=1
                elif mmap[mov_x][mov_y] == 'B': # 변경 
                    blues +=1
    # 노드 방문 탈출: 색 변경 
    if reds <= blues: # 파랑이더 많으면 빨강 제거
        for i in range(N):
            for j in range(M):
                if visited[i][j] == True and mmap[i][j] == 'A':
                    mmap[i][j] ='O'
    else: # 빨강이 더 많으면  파랑 제거 
        for i in range(N):
            for j in range(M):
                if visited[i][j] == True and mmap[i][j] == 'B':
                    mmap[i][j] ='O'        
            
N ,M = map(int,input().split())
mmap = [list(input().strip()) for _ in range(N)]

visited = [[False]*M for _ in range(N)]

for i in range(N):
    for j in range(M):
        if visited[i][j] == False and mmap[i][j] in ('A','B'):
            BFS(i,j,mmap,visited,N,M)
            
# 전부 순회 후 빨간색과 파란색 개수 세기             
R_cnt = 0
B_cnt = 0
for i in range(N):
    for j in range(M):
        if mmap[i][j] =='A':
            R_cnt += 1 
        elif  mmap[i][j] =='B':
            B_cnt += 1   
            
            
print(R_cnt ,B_cnt)         
        

        
        
```

```python
from collections import deque

def BFS(x,y, mmap,visited,N,M):
    
    dx = (0,1,0,-1)
    dy = (1,0,-1,0)
    queue = deque()
    queue.append((x,y))
    visited[x][y]= True
    
    # 빨간색, 파란색 어떤게 더 많을까? 
    reds = 0
    blues = 0
    if mmap[x][y] == 'A':
        reds +=1
    elif mmap[x][y] == 'B': # 변경
        blues+=1
    
    # 방문한 좌표를 따로 저장할 리스트
    visited_cells = []
    visited_cells.append((x,y))
    while queue:
        nx , ny = queue.popleft()
        
        for i in range(4):
            mov_x = nx +dx[i]
            mov_y = ny +dy[i]
            if 0<= mov_x <N and 0<=mov_y <M and visited[mov_x][mov_y]==False and mmap[mov_x][mov_y] in ('A','B'):
                queue.append((mov_x,mov_y))
                visited[mov_x][mov_y] =True
                visited_cells.append((mov_x,mov_y))
                if mmap[mov_x][mov_y] == 'A':
                    reds +=1
                elif mmap[mov_x][mov_y] == 'B': # 변경 
                    blues +=1
    # 규칙 적용: 파란색이 더 많거나 같으면 빨간색을 지우기, 빨간색이 더 많으면 파란색을 지우기
    if blues >= reds:
        # 방문한 좌표들 중에서 빨간색 칸을 빈 칸(O)으로 바꾸기
        for x, y in visited_cells:
            if mmap[x][y] == 'A':
                mmap[x][y] = 'O'
    else:
        # 방문한 좌표들 중에서 파란색 칸을 빈 칸(O)으로 바꾸기
        for x, y in visited_cells:
            if mmap[x][y] == 'B':
                mmap[x][y] = 'O'       
            
N ,M = map(int,input().split())
mmap = [list(input().strip()) for _ in range(N)]

visited = [[False]*M for _ in range(N)]

for i in range(N):
    for j in range(M):
        if visited[i][j] == False and mmap[i][j] in ('A','B'):
            BFS(i,j,mmap,visited,N,M)
            
# 전부 순회 후 빨간색과 파란색 개수 세기             
R_cnt = 0
B_cnt = 0
for i in range(N):
    for j in range(M):
        if mmap[i][j] =='A':
            R_cnt += 1 
        elif  mmap[i][j] =='B':
            B_cnt += 1   
            
            
print(R_cnt ,B_cnt)         
        

        
        
```
