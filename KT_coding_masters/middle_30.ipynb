출력값 설명

만들 수 있는 모든 암호를 사전순으로 출력합니다.

```python
import itertools

# 입력
N, K = map(int, input().split())
chars = sorted(input().strip())  # 입력된 문자를 정렬하고 나서 순열 생성 

# 가능한 모든 순열 생성
combinations = itertools.product(chars, repeat=K)

# 순열 출력
for combination in combinations:
    print(''.join(combination))
```
