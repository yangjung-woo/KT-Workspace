# 소수 구하기
문제설명
소수 정리'란 어떤 자연수 이하의 소수가 몇 개나 있는지 그 값을 어림할 수 있는 정리입니다. 이 정리가 맞는지 의구심이 들은 진수는 직접 확인해 보기로 했습니다. 한참을 계산하던 진수는 소수의 개수를 헤아리는데 몹시 많은 시간일 든다는 걸 깨닫고 여러분들께 도움을 구했습니다. 진수를 위해 자연수 N 이하의 소수 개수를 출력하는 프로그램을 만들어주세요. 이때 소수란 양의 약수가 자기 자신과 1만 존재하는 수를 말합니다. 또한 어떤 정수 A가 어떤 정수 B의 약수라는 것은 B를 A로 나누었을 때 나누어떨어짐을 의미합니다.

입출력 예
입력 #1

978

입력 #2

99

입력값 설명

첫 번째 줄에 1 이상 1000 이하의 자연수 N이 주어집니다.

출력 #1

165

출력 #2

25

출력값 설명

1 이상 N 이하의 자연수 중에서의 소수의 개수를 출력합니다.

```python
# 소수의 개수 구하는 문제 , 1< N< 1000
# 소수인가? 를 확인하는 반복문을 구현할 수 있지만 시간복잡도가 걱정 O(n**2)
# 아리스토테네스의 체 정리를 기반으로  N이 소수인지 확인하기 위해선 sqrt(n) 까지만 확인해도 소수인지 확인할 수 있다
import math
def ispirme(n): # O(log n)
    flag = True
    sq_n = int(math.sqrt(n))
    print(sq_n)
    for i in range(2,sq_n+1):  # 2~ sqrt(n) 까지
        if n % i ==0: # 소수가 아님
            flag = False
            break
    return flag
# N 까지 모든 수 
n= int(input())
cnt =0
for i in range(2,n+1):
    if ispirme(i):  # 평균 시간복잡도 (log N) 
        cnt+=1

print(cnt)
    
```
