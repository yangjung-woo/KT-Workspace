# 콘웨이의 생명 게임

문제설명
콘웨이의 생명 게임이란 격자판의 셀 각각에 위치한 세포의 변화를 관찰하는 게임입니다.
세포는 한 세대가 지나면 죽거나 살아나게 되는데 인접한 칸의 상황에 따라 양상이 달라집니다.
이때 어떤 칸의 인접한 칸이란 다음과 같습니다. 
편의상 어떤 칸이 격자판의 r행, c열에 위치할 때 이를 (r, c) 로 나타냅시다. 그러면 그 격자칸 (r, c)의 인접한 칸은 다음 8칸을 의미합니다.
(r-1, c-1), (r-1, c), (r-1, c+1), (r, c-1), (r, c+1), (r+1, c-1), (r+1, c), (r+1, c+1) 

한편, 한 세대가 지날 때 세포의 변화양상은 다음과 같습니다.
i) 어떤 세포가 살아있는 상태일 때
인접한 칸 중 살아있는 세포가 2개 이거나 3개면 그 세포는 다음 세대에도 생존합니다.
그 외의 경우는 다음 세대에 죽습니다.

ii) 어떤 세포가 죽어있는 상태일 때
인접한 칸 중 살아있는 세포가 3개일 경우만 다음 세대에 그 세포가 살아납니다.
그 외의 경우는 다음 세대에도 여전히 죽어있는 상태입니다.

5 x 5 의 콘웨이의 생명게임의 초기상태가 주어질 때 N번째 세대를 출력하세요.

둘째줄부터 5개 줄에 걸쳐 생명게임의 초기상태가 주어집니다.
0은 죽어있는 세포를, 1은 살아있는 세포를 의미합니다.

출력값 설명

5 x 5 의 콘웨이의 생명게임의 초기상태가 주어질 때 N 번째 세대를 출력하세요.

```python
def get_next_generation(current): 
    next_gen = [[0] * 5 for _ in range(5)] 
    for r in range(5): 
        for c in range(5): 
            live_neighbors = 0 
            # Check all 8 possible neighbors 
            for dr in [-1, 0, 1]: 
                for dc in [-1, 0, 1]: 
                    if dr == 0 and dc == 0: 
                        continue 
                    nr, nc = r + dr, c + dc 
                    if 0 <= nr < 5 and 0 <= nc < 5: 
                        live_neighbors += current[nr][nc] 
             
            if current[r][c] == 1:  # Currently alive 
                if live_neighbors in [2, 3]: 
                    next_gen[r][c] = 1 
                else: 
                    next_gen[r][c] = 0 
            else:  # Currently dead 
                if live_neighbors == 3: 
                    next_gen[r][c] = 1 
                else: 
                    next_gen[r][c] = 0 
                     
    return next_gen 
 
def print_generation(gen): 
    for row in gen: 
        print(''.join(map(str, row))) 
 
#  
N = int(input()) 
 
# Read the initial state 
initial_state = [list(map(int, input().strip())) for _ in range(5)] # 맵은 5행이 최대 
 
# current_generation = initial_state 
 
# Simulate N generations 
for _ in range(N): 
    initial_state = get_next_generation(initial_state) 
 
# Print the result 
print_generation(initial_state) 
```

접근 방법:


1. 5x5 크기의 초기 상태를 입력받습니다. 각 셀은 0(죽은 상태) 또는 1(살아있는 상태)로 주어집니다.

2. 이웃 셀 계산:

각 셀의 주변 8개의 이웃 셀들을 확인하여 살아있는 이웃 셀의 수를 계산합니다.
이웃 셀은 상하좌우 및 대각선 방향에 있는 셀들을 포함하며, 그리드를 벗어나지 않도록 경계 조건을 체크합니다.
3. 세대 업데이트 규칙:

현재 세대에서 셀의 상태는 다음과 같은 규칙을 따릅니다:
현재 살아있는 셀(1):
살아있는 이웃이 2개 또는 3개이면 다음 세대에서도 살아남습니다.
그렇지 않으면 죽습니다.
현재 죽어있는 셀(0):
살아있는 이웃이 정확히 3개인 경우 부활합니다.
그렇지 않으면 죽은 상태를 유지합니다.

4. N 세대 진행:

주어진 세대 수 N만큼 다음 세대를 계산하여 상태를 갱신합니다.

- 핵심 개념:
이웃 셀 확인: 8방향의 셀을 체크하고, 그리드를 벗어나지 않도록 조건을 설정하는 것이 중요합니다.
생명 게임 규칙: 각각의 셀이 살아남거나 죽는 조건을 정확하게 구현하는 것이 문제 해결의 핵심입니다.
이 코드는 일정한 규칙에 따라 그리드를 업데이트하는 시뮬레이션 문제에 속하며, 이웃 셀을 기준으로 각 셀의 상태 변화를 확인하는 방식으로 접근했습니다.

```python
#문제접근 5x5 맵에서 다음세대에 생존할지 아닐지 여부를 반영하는 문제 N번 

# 8방향 
# if (x,y) ==1 일때 
#   인접 좌표들 중에서  1의 개수가 2 or 3이면 >>1 
#   아니면                                  >> 0

# else (x,y)==0 일때
#   인접 좌표들 중에서 1의 개수가 3이면  >>1
#   아니면                             >>0


def step_generation(current):
    dy = [0,-1,-1,-1,0,1,1,1]
    dx = [1,1,0,-1,-1,-1,0,1]
    next_gen = [[0] * 5 for _ in range(5)] 
    # 주위 1의 개수를 확인 
    for x in range(5):
        for y in range(5):
            live_neighbor = 0
             # 주위 1의 개수를 확인
            for d in range(8):
                nx = x +dx[d]
                ny = y+ dy[d]
                if 0<=nx<5 and 0<=ny<5:
                    live_neighbor += current[nx][ny]
                    
            if current[x][y]==1:
                if live_neighbor ==2 or live_neighbor ==3:
                    next_gen[x][y] =1
                else:
                    next_gen[x][y] =0
            
            else: # 0일때 
                if  live_neighbor ==3:
                    next_gen[x][y] =1
                else:
                    next_gen[x][y] =0
    return next_gen        
            
N = int(input())

init_state = [list(map(int,input().strip()))  for _ in range(5)] # 5x5 최대 

for _ in range(N):
    step_state = step_generation(init_state) 
    
print_generation(step_state) 
```
