# 컴퓨터 학원

문제설명
동빈이는 컴퓨터 선생님이 되고 싶었지만 임용고시에 낙방하여 컴퓨터 학원을 차리게 되었습니다. 동빈이네 컴퓨터 학원에서는 시스템 해킹을 가르치는데, 시스템 해킹을 배우는 학생들은 프로그램 취약점을 분석하는 과정에 지쳐서 항상 화가 나 있는 상태입니다.

동빈이네 컴퓨터 학원의 자습실은 2 X N 형태의 직사각형으로 구성되어 있습니다. 자습실 한 칸에는 한 명의 학생만 들어갈 수 있으며, 학생들이 예민하여 가로로든 세로로든 서로 붙어있다면 싸우게 될 확률이 매우 높습니다.

그래서 동빈이는 학생들이 서로 싸우지 않도록 하기 위해 가로와 세로로 한 칸 이상씩 학생들을 떨어뜨려 놓으려고 합니다. 이 때 학생들을 배치할 수 있는 가능한 모든 경우의 수를 출력하세요. 학생들이 한 명도 배치되지 않는 경우도 하나의 경우로 고려할 수 있습니다.

예를 들어 학생이 배치되지 않는 경우를 X라고 하고, 학생이 배치되는 경우를 O라고 표현할 때 2 X 1 크기의 자습실이 있을 때 학생들을 배치하는 경우는

(X, X)

(X, O)

(O, X)으로 총 3가지 경우가 존재합니다.

또한 2 X 2 크기의 자습실이 있을 때 학생들을 배치하는 경우는

(X, X)
(X, X)

(O, X)
(X, X)

(X, O)
(X, X)

(X, X)
(O, X)

(X, X)
(X, O)

(O, X)
(X, O)

(X, O)
(O, X)으로 총 7가지 경우가 존재합니다.

입출력 예
입력 #1

1

입력 #2

2

입력값 설명

첫째 줄에 자습실의 크기 N(1 ≤ N ≤ 10,000)이 주어집니다.

출력 #1

3

출력 #2

7

출력값 설명

첫째 줄에 학생을 배치하는 경우의 수를 796796으로 나눈 나머지를 출력합니다.

# 문제 접근 DP 방식 

- 점화식 구하기 
 dp[n] = dp[n-1]* 2 + dp[n-2]

- 점화식을 어떻게 구하는가??

1. 이전 단계에서  모두 (X X) 를 추가할 수 있음 , dp[n] =dp[n-1]
2. 이전 단계에서  모두 (O X) 또는 (X O)를 반드시 추가할 수 있음 dp[n] + = dp[n-1] + dp[n-1]
3. 이 전전 단계에서  (O X) (X O)  또는 (X O) (O  X)쌍을 반드시 추가할 수 있음   dp[n] +=  dp[n-2]

모든 점화식을 합하면  dp[n] = dp[n-1]*2 + dp[n-2]

```python
n = int(input())
dp = [0]*(n+1)

dp[1] = 3
dp[2] = 7

for i in range(3,n+1):
    dp[i] = dp[i-1]*2 + dp[i-2]
#첫째 줄에 학생을 배치하는 경우의 수를 796796으로 나눈 나머지를 출력합니다.
print(dp[n]%796796)
    

```
