다만 세경이는 게으르기 때문에 조금이라도 산을 내려가면 다시 올라가지 않습니다. 세경이가 산을 내려갈 때는 바로 아래로 내려가거나, 한 칸 오른쪽으로 내려갑니다. 세경이가 산을 다 내려올 때까지 주을 수 있는 동전의 최대 개수를 구하는 프로그램을 작성하세요.

```python
n = int(input())
triangle = []
dp = []

for i in range(n):
    triangle.append(list(map(int,input().split())))
    dp.append([0]*(i+1)) # 층마다 요소 1개씩 증가하도록 
    
dp[0][0] = triangle[0][0] # 최 상단은 단 1개뿐이기에 그대로 적용

for i in range(1,n):
    for j in range(0,i+1):
        # 3가지 경우의 수 중 1. 가장 왼쪽 끝
        if j ==0:
            dp[i][j] = dp[i-1][j] + triangle[i][j]
        elif j ==i: # 2. 가장 오른쪽 끝
            dp[i][j] = dp[i-1][j-1] + triangle[i][j]
        else: # 그외 중간 경우의 수 
            dp[i][j] = max(dp[i-1][j-1] , dp[i-1][j]) + triangle[i][j]

print(max(dp[n-1]))

```

```python
def max_coins(N, triangle):
    # DP 테이블 초기화 (삼각형과 동일한 크기)
    dp = [[0] * (i + 1) for i in range(N)]
    
    # 첫 번째 위치의 동전 수로 DP 시작
    dp[0][0] = triangle[0][0]
    
    # DP 수행
    for i in range(1, N):
        for j in range(i + 1):
            if j == 0:
                # 맨 왼쪽 끝은 위에서 바로 내려오는 경우만 가능
                dp[i][j] = dp[i - 1][j] + triangle[i][j]
            elif j == i:
                # 맨 오른쪽 끝은 바로 위의 오른쪽 대각선에서만 내려올 수 있음
                dp[i][j] = dp[i - 1][j - 1] + triangle[i][j]
            else:
                # 중간에 있는 경우, 위에서 바로 내려오거나 오른쪽 대각선에서 내려오는 경우 중 큰 값 선택
                dp[i][j] = max(dp[i - 1][j - 1], dp[i - 1][j]) + triangle[i][j]
    
    # 마지막 행에서 가장 큰 값이 세경이가 주울 수 있는 최대 동전 수
    return max(dp[N - 1])

# 입력 처리
N = int(input())  # 산의 높이
triangle = [list(map(int, input().split())) for _ in range(N)]

# 최대 동전 수 출력
print(max_coins(N, triangle))
```
