# 슬라이딩 퍼즐
문제설명
슬라이딩 퍼즐을 풀어봅시다.
슬라이딩 퍼즐은 2 x 2의 판에서 각 조각들을 인접한 비어있는 칸으로 밀어서 최종적으로
12
3X
의 형태로 만드는 퍼즐입니다. X는 비어있는 칸을 말합니다.
슬라이딩 퍼즐이 입력으로 주어졌을 때, 퍼즐을 풀기 위한 최소 조각 이동 횟수를 출력하세요.

입출력 예
입력 #1

12
X3

입력 #2

3X
12

입력값 설명

2줄에 걸쳐, 슬라이딩 퍼즐이 주어집니다.

출력 #1

1

출력 #2

-1

출력값 설명

퍼즐을 풀기 위한 최소 조각 이동 횟수를 출력합니다.
만약 불가능하다면, 대신 -1을 출력합니다.

```python
# 문제접근: X 빈칸위치에서 상하좌우 Swap할 수 있는 것은 인접노드 => BFS 탐색으로 해결
# 목표 string = 123X
# 초기상태 2차원 배열을 1차원 스트링으로 변환 = ''.join(start_input)


# dfs(start_string)
# 목표는 123X
# 상하좌우 이동 설정
# 방문할 queue 에는 (시작String, 빈공간위치 , step수 = init :0 )을 저장
# visit = [123x , 31x2 등 가능한 문자열들을 저장 , 단 중복제거 set 자료 ]
# while 
    # popleft
    # if 목표라면 종료
    
    # x, y = divmod(zero_index, 2) # 몫과 나머지를 동시에 구함  ex 3   => (1,1)
    
    # for i in range(4)
    #   nx ,ny =   x + dx[i], y + dy[i]
    #if 0 <= nx < 2 and 0 <= ny < 2:
    #            new_zero_index = nx * 2 + ny #  1~4 위치를 좌표로 변환 , (1,1) => 3 
    #            new_state = list(current)
                # 빈공간 블럭 스왑 
    #            new_state[zero_index], new_state[new_zero_index] = new_state[new_zero_index], new_state[zero_index]
    #            new_state_str = ''.join(new_state)
                
    #            if new_state_str not in visited:
    #                visited.add(new_state_str)
    #                queue.append((new_state_str, new_zero_index, moves + 1))

    # return -1  # 목표 상태에 도달할 수 없는 경우
    

from collections import deque

def sliding_puzzle_BFS(start):
    target_str = '123X'  # 목표 상태
    dy = (0,1,0,-1) # 이동 방향
    dx = (1,0,-1,0)
    # BFS 초기 설정
    queue = deque([(start, start.index('X'), 0)])  # (현재 상태, 빈 공간 인덱스, 이동 횟수)
    visited = set()
    visited.add(start)

    while queue:
        current, zero_index, moves = queue.popleft()
        
        if current == target_str:
            return moves
        
        x, y = divmod(zero_index, 2) # 몫과 나머지를 동시에 구함  ex 3   => (1,1)

        for i in range(4):
            nx, ny = x + dx[i], y + dy[i]
            
            if 0 <= nx < 2 and 0 <= ny < 2:
                new_zero_index = nx * 2 + ny #  1~4 위치를 좌표로 변환 , (1,1) => 3 
                new_state = list(current)
                # 빈공간 블럭 스왑 
                new_state[zero_index], new_state[new_zero_index] = new_state[new_zero_index], new_state[zero_index]
                new_state_str = ''.join(new_state)
                
                if new_state_str not in visited:
                    visited.add(new_state_str)
                    queue.append((new_state_str, new_zero_index, moves + 1))

    return -1  # 목표 상태에 도달할 수 없는 경우

# 입력 처리
start_input = [input().strip() for _ in range(2)]
start_state = ''.join(start_input)
    
result = sliding_puzzle_BFS(start_state)
print(result)
```
