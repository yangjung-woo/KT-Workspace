```python
n = int(input())
M = []
dp = []
for i in range(n):
    M.append(list(map(int,input().split())) )
    dp.append([9999999]*n)
    
dp[0][0] = M[0][0]

for i in range(n):
    for j in range(n):
        # 총 5가지 경우의 수 : 모서리 4개 , 그외 1개 
        if i ==0:
            if j == 0:
                dp[i][j] = M[0][0]
            elif j ==n-1: # 좌 하 
                dp[i][j]  = min(dp[i+1][j] , dp[i][j-1]) + M[i][j]
            else: # 좌 우 하
                dp[i][j]  = min(dp[i][j+1],dp[i+1][j] , dp[i][j-1]) + M[i][j]
                
        elif j ==0:
            if i == 0:
                dp[i][j] = M[0][0]
            elif i ==n-1: # 상  우 
                dp[i][j]  = min(dp[i-1][j] , dp[i][j+1]) + M[i][j]
            else: # 상 하 우
                dp[i][j]  = min(dp[i-1][j],dp[i+1][j] , dp[i][j+1]) + M[i][j]
            
        elif i == n-1:
            if j == 0: # 상 우 
                dp[i][j]  = min(dp[i-1][j] , dp[i][j+1]) + M[i][j]
            elif j == n-1: # 상좌 
                dp[i][j]  = min(dp[i-1][j] , dp[i][j-1]) + M[i][j]
            else: # 좌 우 상
                dp[i][j]  = min(dp[i][j-1],dp[i][j+1] , dp[i-1][j]) + M[i][j]
        
        elif j == n-1:
            if i == 0: # 좌 하 
                dp[i][j]  = min(dp[i+1][j] , dp[i][j-1]) + M[i][j]
            elif i == n-1: # 상 좌
                dp[i][j]  = min(dp[i-1][j] , dp[i][j-1]) + M[i][j]
            else: # 좌 상 하 
                dp[i][j]  = min(dp[i][j-1],dp[i+1][j] , dp[i+1][j]) + M[i][j]
            
        else:
            dp[i][j] = min(dp[i][j-1],dp[i][j+1],dp[i+1][j] , dp[i+1][j]) + M[i][j]


            
print(dp)
            
            
```

# DP 실패

DP + BFS 합쳐진 고난이도 문제 


```python
from collections import deque

def min_rice_cakes_bfs_dp(N, board):
    directions = [(1, 0), (-1, 0), (0, 1), (0, -1)]
    queue = deque([(0, 0)])  # (행, 열)
    min_rice_cakes = [[9999999] * N for _ in range(N)]
    min_rice_cakes[0][0] = board[0][0]

    while queue:
        x, y = queue.popleft()

        for dx, dy in directions:
            nx, ny = x + dx, y + dy

            if 0 <= nx < N and 0 <= ny < N:
                new_cakes = min_rice_cakes[x][y] + board[nx][ny]
                if new_cakes < min_rice_cakes[nx][ny]:# 최단거리일 경우 갱신 
                    min_rice_cakes[nx][ny] = new_cakes
                    queue.append((nx, ny))

    return min_rice_cakes[N-1][N-1]

# 입력 예제
N = int(input())
board = []

for i in range(N):
    board.append(list(map(int,input().split())) )
    
print(min_rice_cakes_bfs_dp(N, board))
```

```python
from collections import deque


def min_rice_cakes_bfs_dp(N,board):
    direction = [(1, 0), (-1, 0), (0, 1), (0, -1)]
    queue = deque([(0,0)]) # 방문 할 노드 
    
    dp = [[9999999] * N for _ in range(N)]
    dp[0][0] = board[0][0]
    
    while queue:
        
        x, y = queue.popleft()
        
        for dx ,dy in direction:
            nx = dx + x
            ny = dy + y
            if 0<=nx<N  and  0<=ny<N:
                new_cost = dp[x][y]+board[nx][ny] 
                if new_cost < dp[nx][ny]: # 최단 거리만 queue에 저장 
                    dp[nx][ny] = new_cost
                    queue.append((nx,ny))
    return dp
            
            
    


N = int(input())
board = []

for i in range(N):
    board.append(list(map(int,input().split())) )

print(min_rice_cakes_bfs_dp(N, board))
```
