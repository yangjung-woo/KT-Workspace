# 팰린드롬 만들기

문제설명
팰린드롬은 역순으로 읽어도 같은 낱말을 뜻합니다. 예를 들어, EYE, MADAM은 팰린드롬이지만 APPLE, CODE는 아닙니다.
여러분에게 문자열들이 주어집니다. 이 문자열들을 적절히 재배열해서, 하나의 팰린드롬 문자열을 만들 수 있는지 여부를 출력하는 프로그램을 작성해 주세요.

입출력 예
입력 #1

3

AB

CBA

C

입력 #2

3

A

CBA

CB

입력값 설명

첫째 줄에 문자열의 수 N이 주어집니다. (1 ≤ N ≤ 8)
이어서 N개의 줄에, 알파벳 대문자만으로 이루어진 문자열이 주어집니다. (1 ≤ 문자열의 길이 ≤ 20)

출력 #1

YES

출력 #2

NO

출력값 설명

문자열들을 적절히 재배열한 후 하나의 문자열로 합쳤을 때, 팰린드롬으로 만들 수 있으면 YES, 그렇지 않으면 NO를 출력합니다.
