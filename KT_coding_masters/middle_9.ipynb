# 리버스 게임
문제설명
리버스라는 보드 게임이 있습니다. 게임 규칙은 이러합니다. 두 명이서 가위바위보를 하고, 이긴 사람은 바둑판에 앞면은 검은색, 뒷면은 흰색인 바둑들을 임의로 깔아놓습니다.

가위바위보에서 진 사람은 바둑돌을 뒤집을 수 있는데, 한번 뒤집을 때 해당 열이나 행의 모든 바둑돌을 뒤집어야 합니다. 예를 들어 3 X 3 형태로 바둑돌이 놓여져 있을 때 원하는 바둑돌 하나만 뒤집을 수는 없고, 특정한 열이나 행에 해당하는 바둑돌 3개를 모두 뒤집어야 합니다. 게임에서 이기기 위해서는 원하는 만큼 바둑돌을 뒤집어서 위에서 보이는 흰색 바둑돌의 개수가 최소가 되어야 합니다.

자연수 N이 주어지고, N X N 형태의 바둑돌들이 주어질 때 바둑돌을 뒤집는 사람이 만들 수 있는 위에서 보이는 흰색 바둑돌의 최소 개수를 출력하는 프로그램을 만드세요.

입출력 예
입력 #1

3

BBW

WBB

WBW

입력 #2

2

BW

WB

입력값 설명

첫째 줄에 자연수 N이 주어집니다. (1 ≤ N ≤ 10)
둘째 줄부터 N개의 줄에 걸쳐 바둑돌들의 상태가 길이가 N인 문자열로 주어집니다.
i번째 문자열의 j번째 문자가 'B'인 경우 i행 j열의 바둑돌의 윗면이 검은색이고, 'W'인 경우 윗면이 흰색이라는 의미입니다. 

출력 #1

2

출력 #2

0

출력값 설명

한 행 또는 한 열에 놓인 N개의 바둑돌을 모두 뒤집는 작업들을 무한정 수행할 수 있다고 했을 때, 윗면이 흰색이 되는 바둑돌의 최소 개수를 출력합니다.

# 문제 접근 방식

완전 탐색 

모든 행, 열을 뒤집는 경우의 수를 전부 파악해서 최소 바둑돌 개수를 파악

왜? 완전탐색이 가능? >> 1<n<10   즉 모든 행,열 의 경우의 수 (2**10)*(2**10)  >> 10**6 개  시간초과가 발생하지 않음


+ 유용한 메소드 import itertools import product

 itertools에는 여러가지 함수들이 많이 있는데, 그 중 조합형 이터레이터인

- combinations() # 순열

- combinations_with_replacement()

- product():  여러 iterable의 데카르트곱 리턴
product는 다른 함수와 달리 인자로 여러 iterable을 넣어줄 수 있고 그 친구들간의 모든 짝을 지어서 리턴합니다. 

- permutations()

에 대해 알아봅시다.

```python
from itertools import product


# 바둑돌을 뒤집는 함수 (B <-> W)
def flip(board, N, rows, cols):
    new_board = [list(row) for row in board]  # 원본 배열 복사
    for r in range(N):
        if rows[r]:  # r번째 행을 뒤집음
            for c in range(N):
                if board[r][c] == 'W':
                    new_board[r][c] = 'B' 
                else:
                    new_board[r][c] ='W'
    for c in range(N):
        if cols[c]:  # c번째 열을 뒤집음
            for r in range(N):
                if new_board[r][c] == 'W':
                    new_board[r][c] = 'B' 
                else:
                    new_board[r][c] ='W'
    return new_board

# 흰색 바둑돌(W)의 개수를 세는 함수
def count_white(board):
    return sum(row.count('W') for row in board)

# 입력 처리
n = int(input())
board = []
for i in range(n):
    board.append(list(input().strip()))
    
# 모든 행 열 뒤집기 경우의 수 
min_white = n*n

for row in product([0,1], repeat=n):
    for col in product([0,1], repeat=n):
        # 판 행 열 뒤집기 
        flipped_board = flip(board, n, row, col)
        # 흰 돌 세기
        white_count = count_white(flipped_board)
        # 최소값 갱신
        min_white = min(min_white, white_count)
        
print(min_white)
        
        
```

```python
from itertools import product

# 바둑돌을 뒤집는 함수 (B <-> W)
def flip(board, N, rows, cols):
    new_board = [list(row) for row in board]  # 원본 배열 복사
    for r in range(N):
        if rows[r]:  # r번째 행을 뒤집음
            for c in range(N):
                new_board[r][c] = 'B' if board[r][c] == 'W' else 'W'
    for c in range(N):
        if cols[c]:  # c번째 열을 뒤집음
            for r in range(N):
                new_board[r][c] = 'B' if new_board[r][c] == 'W' else 'W'
    return new_board

# 흰색 바둑돌(W)의 개수를 세는 함수
def count_white(board):
    return sum(row.count('W') for row in board)

def min_white_stones(N, board):
    min_white = N * N  # 흰색 바둑돌의 최대 개수
    # 모든 행, 열의 뒤집기 여부 경우의 수를 탐색
    for rows in product([0, 1], repeat=N):
        for cols in product([0, 1], repeat=N):
            flipped_board = flip(board, N, rows, cols)
            white_count = count_white(flipped_board)
            min_white = min(min_white, white_count)
    return min_white

# 입력 처리
N = int(input())
board = [list(input().strip()) for _ in range(N)]

# 결과 출력
print(min_white_stones(N, board))
```

```python
# 바둑돌을 뒤집는 함수 (B <-> W)
def flip_row(board, row):
    return ['B' if board[row][c] == 'W' else 'W' for c in range(len(board))]

def flip_col(board, col):
    for r in range(len(board)):
        board[r][col] = 'B' if board[r][col] == 'W' else 'W'
    return board

# 흰색 바둑돌(W)의 개수를 세는 함수
def count_white(board):
    return sum(row.count('W') for row in board)

# 재귀적으로 모든 경우의 수 탐색
def dfs(board, row, N):
    if row == N:
        # 모든 행에 대해 뒤집기를 결정한 후 흰색 바둑돌 개수 계산
        return count_white(board)

    # 현재 행을 뒤집지 않는 경우
    min_white = dfs([r[:] for r in board], row + 1, N)

    # 현재 행을 뒤집는 경우
    flipped_row_board = [r[:] for r in board]
    flipped_row_board[row] = flip_row(flipped_row_board, row)
    min_white = min(min_white, dfs(flipped_row_board, row + 1, N))

    return min_white

def solve(N, board):
    # 모든 열을 뒤집는 경우의 수를 고려
    min_white = dfs(board, 0, N)
    for col in range(N):
        flipped_board = [r[:] for r in board]
        flip_col(flipped_board, col)  # 열을 뒤집음
        min_white = min(min_white, dfs(flipped_board, 0, N))

    return min_white

# 입력 처리
N = int(input())
board = [list(input().strip()) for _ in range(N)]

# 결과 출력
print(solve(N, board))
```
